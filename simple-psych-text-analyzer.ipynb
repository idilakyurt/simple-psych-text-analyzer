```python
positive_words = [
    "happy", "joy", "good", "great", "love", "smile",
    "friend", "success", "bright", "hope"
]

negative_words = [
    "sad", "bad", "hate", "cry", "fail", "stress",
    "worry", "down", "fear", "trouble"
]

print("Emotional dictionaries defined!")
```

```python
sample_text = """
I am so happy that I found a new friend,
but I still stress about my final exams.
It feels great to be done with all the worry! Success is near.
"""
clean_text = sample_text.lower()
clean_text = clean_text.replace("!", "").replace(",", "").replace(".", "").replace("\n", " ")
words = clean_text.split()
print("--- Original Text ---")
print(sample_text)
print("\n--- Cleaned and Tokenized Words ---")
print(words)
```

```python
positive_score = 0
negative_score = 0

for word in words:
    if word in positive_words:
        positive_score = positive_score + 1

    elif word in negative_words:
        negative_score = negative_score + 1

print(f"\nAnalysis Complete!")
```

```python
print("--- Text Analyzer Report ---")
print(f"Total Words Analyzed: {len(words)}")

print(f"\nPositive Score: {positive_score} words")
print(f"Negative Score: {negative_score} words")


if positive_score > negative_score:
    print("\nInterpretation: The text exhibits a generally positive tone.")
elif negative_score > positive_score:
    print("\nInterpretation: The text exhibits a generally negative tone.")
else:
    print("\nInterpretation: The text exhibits a neutral or balanced tone.")
```
